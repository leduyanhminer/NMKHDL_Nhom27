```python
import pandas as pd
df_train = pd.read_csv('../data/train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('../data/test.csv')
df_test.head()
```

```python
label_list = ['Game', 'Van Phong', 'Do Hoa', 'Lap Trinh', 'Doanh Nhan']
X_train = df_train.drop(label_list, axis=1)
X_train = X_train.drop(['Unnamed: 0', '_id'], axis = 1)
y_train = df_train[label_list]
X_test = df_test.drop(label_list, axis=1)
X_test = X_test.drop(['Unnamed: 0', '_id'], axis = 1)
y_test = df_test[label_list]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

imputer = SimpleImputer(strategy='mean')
scaler = StandardScaler()
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)  
```

```python
from sklearn.metrics import f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from keras.models import Sequential
from keras.layers import Dense
```

<h3>KNN</h3>

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

predictions = knn.predict(X_test)
f1_macro = f1_score(y_test, predictions, average='macro')
f1_micro = f1_score(y_test, predictions, average='micro')
f1_weighted = f1_score(y_test, predictions, average='weighted')

print("F1 Score (Macro):", f1_macro)
print("F1 Score (Micro):", f1_micro)
print("F1 Score (Weighted):", f1_weighted)
```

<h3>RandomForestClassifier</h3>

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

predictions = rf.predict(X_test)
f1_macro = f1_score(y_test, predictions, average='macro')
f1_micro = f1_score(y_test, predictions, average='micro')
f1_weighted = f1_score(y_test, predictions, average='weighted')

print("F1 Score (Macro):", f1_macro)
print("F1 Score (Micro):", f1_micro)
print("F1 Score (Weighted):", f1_weighted)
```

<h3>Multi-Layer Perceptron</h3>

```python
n_features = X_train.shape[1]
model = Sequential()

model.add(Dense(64, input_dim=n_features, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(5, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
predictions = model.predict(X_test)
predictions_binary = (predictions > 0.5).astype(int)

f1_macro = f1_score(y_test, predictions_binary, average='macro')
f1_micro = f1_score(y_test, predictions_binary, average='micro')
f1_weighted = f1_score(y_test, predictions_binary, average='weighted')

print("F1 Score (Macro):", f1_macro)
print("F1 Score (Micro):", f1_micro)
print("F1 Score (Weighted):", f1_weighted)
```
