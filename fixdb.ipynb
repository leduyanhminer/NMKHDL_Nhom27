```python
import csv

updates = []

with open('laptop_new2.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        # try:
        #     ram = int(row["Memory (RAM)"])
        # except ValueError:
        #     ram = None 
        # try:
        #     screen_size = float(row["Screen size"])
        # except ValueError:
        #     screen_size = None
        # try:
        #     price = float(row["Price"])
        # except ValueError:
        #     price = None
        try:
            cpu_score = int(row["CPU Score"])
        except ValueError:
            cpu_score = 0
        try:
            gpu_score = int(row["GPU Score"])
        except ValueError:
            gpu_score = 0

        update = {
            "_id": row["_id"],
            # "Memory (RAM)": ram,
            # "Screen size": screen_size,
            # "Price": price
            "CPU Score": cpu_score,
            "GPU Score": gpu_score
        }
        updates.append(update)
```

```python
print(updates[5:10])
```

```python
from pymongo import MongoClient
from config import *
from bson import ObjectId

client = MongoClient(DB_URI)
db = client[DB_NAME]
collection = db[DB_COLLECTION_NAME]

for update in updates:
    collection.update_one(
        {'_id': ObjectId(update['_id'])},
        {'$set': {
            # 'Memory (RAM)': update['Memory (RAM)'],
            # 'Screen size': update['Screen size'],
            # "Price": update["Price"]
            'CPU Score': update['CPU Score'],
            'GPU Score': update['GPU Score']
        }}
    )
```
